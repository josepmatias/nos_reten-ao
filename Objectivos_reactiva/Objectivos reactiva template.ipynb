```python
#visualizar folha objectivos reactiva no resumo_template
resumo_template=pd.read_excel(resumo_template_nome, sheet_name="Objetivos reativa")

resumo_template.head()
```

```python
#visualizar objectivos reactiva para o novo mês
objectivo_reactiva=pd.read_excel(objectivos_reactiva_nome, sheet_name=0)

objectivo_reactiva.head()
```

```python
#definir o novo período
periodo=objectivo_reactiva["Descritivo_IB"].unique().tolist() [0].split("_")[0]

periodo_antigo=resumo_template["Objective Name"].unique().tolist() [0].split("_") [2]


print(f"periodo actual: {periodo} e período antigo: {periodo_antigo}")


resumo_template["Objective Name"]=resumo_template["Objective Name"].str.replace(periodo_antigo, periodo)
resumo_template["Objective Name.1"]=resumo_template["Objective Name.1"].str.replace(periodo_antigo, periodo)

resumo_template2=resumo_template.copy()
resumo_template2.head()
```

```python
#actualizar o campo das datas: "Begin Date" e "End Date"

resumo_template2["Begin Date"]=start_date
resumo_template2["End Date"]=end_date

resumo_template2["Begin Date"]=pd.to_datetime(resumo_template2["Begin Date"]).dt.strftime('%d/%m/%Y')

resumo_template2["End Date"]=pd.to_datetime(resumo_template2["End Date"]).dt.strftime('%d/%m/%Y')


resumo_template2[["Begin Date", "End Date"]]
```

```python

#fazer transformações na dataframe resumo_template - modificar "EXCLUSIVO", excluir linhas "decisor" e substituir "RRR_GCO_"

def transformacao (df, col="Objective Name"):
  df[col]=df[col].str.replace("EXCLUSIVOS", "Exclusivos")
  df[col]=df[col].apply(lambda i: str(i).replace("RRR_GCO_", "").strip())

  return df

transformacao(resumo_template2)

dicionario_mat={"MATRIZ BASE 2LO": "AT_TM 2LO", "MATRIZ BASE MTR": "AT_TM MTR", "MATRIZ BASE IN": "AT_TM IN",\
"MATRIZ BASE OUT": "AT_TM OUT"}



for i, j in dicionario_mat.items():
  indice=resumo_template2.loc[resumo_template2["Objective Name"].str.find(j)!=-1, "Objective Name"].index
  resumo_template2.loc[indice, "Objective Name"]=resumo_template2.loc[resumo_template2["Objective Name"].str.find(j)!=-1, "Objective Name"].str.replace(j, i)


#verificar quais são as linhas em comum entre resumo_template2 e objectivo_reactiva2

lista_comum=[i for i in resumo_template2["Objective Name"].tolist() if i in objectivo_reactiva["Descritivo_IB"].tolist()]

lista_val_comum=objectivo_reactiva.loc[objectivo_reactiva["Descritivo_IB"].isin(lista_comum), "Objetivo"].tolist()


#mudar valores comuns na dataframe resumo_template2
for i, j in zip(lista_comum, lista_val_comum):
  indice=resumo_template2.loc[resumo_template2["Objective Name"]==i, "Objective Name"].index
  resumo_template2.loc[indice, "Objective\n(Value)"]=j
```

```python
#substituir string "---" por 0

resumo_template2["Objective\n(Value)"]=resumo_template2["Objective\n(Value)"].replace("---", 0)

#arredondar a coluna "Objective\n(Value)" para 2 casas decimais

resumo_template2["Objective\n(Value)"]=resumo_template2["Objective\n(Value)"].apply(lambda i: round(i, 2))
```

```python
#formatar tabela - passos finais
resumo_template2=resumo_template2[resumo_template2.iloc[:, 0].isna()==False]

resumo_template2["Objective Name"]=resumo_template2["Objective Name"].apply(lambda i: "RRR_GCO_" + i)
```

```python
#retomar os labels correctos
resumo_template2["Objective Name"]=resumo_template["Objective Name"]
```
